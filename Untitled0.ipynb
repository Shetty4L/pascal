```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
val_data = pd.read_table('val_accuracies.log',delimiter=',')
```

```python
plt.figure()
plt.title('Plot of Validation Accuracies vs Number of Iterations')
plt.ylabel('Validation Accuracies')
plt.xlabel('Number of Iterations')
m1, = plt.plot(np.arange(0,10001,1000),np.array(val_data['accuracy1']),label='Model 1')
m2, = plt.plot(np.arange(0,10001,1000),np.array(val_data['accuracy2']),label='Model 2')
m3, = plt.plot(np.arange(0,10001,1000),np.array(val_data['accuracy3']),label='Model 3')
m4, = plt.plot(np.arange(0,10001,1000),np.array(val_data['accuracy4']),label='Model 4')
m5, = plt.plot(np.arange(0,10001,1000),np.array(val_data['accuracy5']),label='Model 5')
plt.legend(handles=[m1,m2,m3,m4,m5],loc=4)
plt.savefig('val_acc.png')
plt.show()
```
